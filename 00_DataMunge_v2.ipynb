```python
import sqlite3
import string
# import re
import numpy as np
import pandas as pd
# import patsy as pt

pd.set_option('display.mpl_style', 'default')
pd.set_option('display.notebook_repr_html', True)
np.random.seed(0)
```

```python
cnx = sqlite3.connect('data/volunteerireland_datadive2015.db')
```

# Read Data

## Opps all time

```python
dfoppsaltm = pd.read_excel('data/df_oppsaltm.xlsx') #dtype={'GEOGID':str,'GEOGTYPE':str,'GEOGDESC':str})
dfoppsaltm.rename(columns=lambda x: 'altm_{}'.format('_'.join([y.strip(string.punctuation) for y in x.lower().split()])), inplace=True)
dfoppsaltm['altm_id'] = np.arange(dfoppsaltm.shape[0])
print(dfoppsaltm.shape)
dfoppsaltm.head()
```

## Orgs

```python
dforgs = pd.read_excel('data/df_orgs.xlsx')
dforgs.rename(columns=lambda x: 'orgs_{}'.format('_'.join([y.strip(string.punctuation) for y in x.lower().split()])), inplace=True)
dforgs['orgs_id'] = np.arange(dforgs.shape[0])
print(dforgs.shape)
dforgs.head()
```

## Opps applied for

```python
dfoppsapfr = pd.read_excel('data/df_oppsapfr.xlsx')
dfoppsapfr.rename(columns=lambda x: 'apfr_{}'.format('_'.join([y.strip(string.punctuation) for y in x.lower().split()])), inplace=True)
dfoppsapfr['apfr_id'] = np.arange(dfoppsapfr.shape[0])
print(dfoppsapfr.shape)
dfoppsapfr.head()
```

## Opps Placed

```python
dfplc = pd.read_excel('data/df_plc.xlsx')
dfplc.rename(columns=lambda x: 'plc_{}'.format('_'.join([y.strip(string.punctuation) for y in x.lower().split()])), inplace=True)
dfplc['plc_id'] = np.arange(dfplc.shape[0])
print(dfplc.shape)
dfplc.head()
```

## Volunteers

```python
dfvol = pd.read_excel('data/df_vol.xlsx')
dfvol.rename(columns=lambda x: 'vol_{}'.format('_'.join([y.strip(string.punctuation) for y in x.lower().split()])), inplace=True)
dfvol['vol_id'] = np.arange(dfvol.shape[0])
print(dfvol.shape)
dfvol.head()
```

# Join some stuuf

```python
df1 = pd.merge(dfoppsaltm[['altm_id','altm_organisation_id','altm_volunteer_opportunity_id']]
               ,dforgs[['orgs_id','orgs_organisation_id']]
               ,how='left', left_on='altm_organisation_id', right_on='orgs_organisation_id')
print(df1.shape)
df1.head().T
```

```python
df2 = pd.merge(df1
               ,dfoppsapfr[['apfr_id','apfr_vop_name','apfr_volunteer_volunteer_id']]
               , how='left', left_on='altm_volunteer_opportunity_id', right_on='apfr_vop_name')
print(df2.shape)
df2.head().T
```

```python
dfvol_apfr = dfvol.copy()
dfvol_apfr.rename(columns={'vol_id':'vol_id_apfr','vol_volunteer_id':'vol_volunteer_id_apfr'}, inplace=True)

df3 = pd.merge(df2
               ,dfvol_apfr[['vol_id_apfr','vol_volunteer_id_apfr']]
               ,how='left', left_on='apfr_volunteer_volunteer_id', right_on='vol_volunteer_id_apfr')
print(df3.shape)
df3.head().T
```

```python
df4 = pd.merge(df3
               , dfplc[['plc_id','plc_volunteer_opportunity_record_id','plc_volunteer_id']]
               , how='left', left_on='apfr_vop_name', right_on='plc_volunteer_opportunity_record_id')
print(df4.shape)
df4.head().T
```

```python
dfvol_plc = dfvol.copy()
dfvol_plc.rename(columns={'vol_id':'vol_id_plc','vol_volunteer_id':'vol_volunteer_id_plc'}, inplace=True)

df5 = pd.merge(df4
               ,dfvol_plc[['vol_id_plc','vol_volunteer_id_plc']]
               ,how='left', left_on='plc_volunteer_id', right_on='vol_volunteer_id_plc')
print(df5.shape)
df5.head().T
```

# Write to DB

## Original Tables

```python
dfoppsaltm.to_sql('dfoppsaltm', cnx, if_exists='replace')
dforgs.to_sql('dforgs', cnx, if_exists='replace')
dfoppsapfr.to_sql('dfoppsapfr', cnx, if_exists='replace')
dfplc.to_sql('dfplc', cnx, if_exists='replace')
dfvol.to_sql('dfvol', cnx, if_exists='replace')
```

## Merged Table

```python
df5.to_sql('vi_merged', cnx, if_exists='replace')
```
